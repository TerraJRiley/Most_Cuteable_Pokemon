```python
dict_pokemon = {}
for pokemon in list_dex:
    try:
        num_gender_rate = requests.get("https://pokeapi.co/api/v2/pokemon-species/" + pokemon.lower()).json()["gender_rate"]
        json_encounters = requests.get("https://pokeapi.co/api/v2/pokemon/" + pokemon.lower() + "/encounters").json()
        dict_pokemon[pokemon] = {"Gender Rate" : num_gender_rate,
                                 "Encounters" : json_encounters}
    except:
        print(pokemon, " was skipped.")
```

```python
list_dicts = []
for pokemon in dict_pokemon.keys():                           # iterate pokemon in dictionary
    #print(pokemon)
    #print("-")
    for dict_one in dict_pokemon[pokemon]["Encounters"]: # iterate dict of encounter values
        #print(dict_one)
        #print("--")
        for dict_two in dict_one["version_details"]:        # iterate dict of version details
            #print(dict_two)
            #print("---")
            str_version = dict_two['version']["name"]
            #print(str_version)
            if (str_version == "pearl") or (str_version == "soulsilver"):
                #print(str_version)
                #print("if")
                for dict_tre in dict_two["encounter_details"]:
                    #print(dict_tre)
                    #print("----")
                    list_dicts.append({
                        "Pokemon"   : pokemon,
                        "Gender"    : dict_pokemon[pokemon]["Gender Rate"],
                        "Location"  : dict_one["location_area"]["name"],
                        "Chance"    : dict_tre["chance"],
                        "Max Chance": dict_two['max_chance'],
                        "Version"   : dict_two['version']["name"],
                        #"Condition" : dict_tre["condition_values"][0]["name"],
                        "Method"    : dict_tre["method"]["name"]})
                    try:
                        list_dicts[-1]["Condition"] = dict_tre["condition_values"][0]["name"]
                    except:
                        pass
```
